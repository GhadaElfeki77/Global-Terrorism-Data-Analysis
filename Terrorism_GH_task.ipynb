```python
for col in df.columns:
  pct_missing = np.mean(df[col].isnull())
  print('{}  -  {}%'.format(col, round(pct_missing*100)))
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    pct_missing =  round(pct_missing*100)
    if pct_missing >= 30 :
      df.drop([col], axis = 1, inplace = True)
      print('Droping {} Column as it contain - {}% null values'.format(col, pct_missing))
```

```python
df.shape
```

```python
for column in df.columns:
    if df[column].dtype == 'object':  # For categorical columns
        df[column] = df[column].fillna(df[column].mode()[0])
    elif df[column].dtype in ['int64', 'float64']:  # For numerical columns
        df[column] = df[column].fillna(df[column].mean())
```

##  **Terrorism via Years**

**terrorist incidents success from 1970 to 2018**

```python
sns.set_theme(rc={'figure.figsize':(20, 10)}, font_scale=1.4)
plt.plot(df.groupby('iyear').sum()['success'],color = 'b',ls = '-',lw = 8, alpha = 0.85)
plt.title(" Terrorist Incidents via years ", size=20 )
# Add  X-axis Label
plt.xlabel('Years',color = 'k',size=16)
# Add Y-axis Label
plt.ylabel('Terrorist Incidents',color = 'k',size=16)
plt.savefig('Terrorist Incidents via years')
plt.show()
```

**Insight: Number of terrorist attacks increased sharply after 2001, with a peak around 2014–2015 mainly due to ISIS activity in Iraq and Syria**

**kills number from 1970 to 2018 becuase of terrorist incidents**

```python
sns.set_theme(rc={'figure.figsize':(20, 10)}, font_scale=1.4)
plt.plot(df.groupby('iyear').sum()['nkill'],color = 'b',ls = '-',lw = 8, alpha = 0.85)
plt.title(" Number of Kills Because of Terrorist Incidents ", size=20 )
# Add  X-axis Label
plt.xlabel('Years',color = 'k',size=16)
# Add Y-axis Label
plt.ylabel('Number of Kills ',color = 'k',size=16)
plt.savefig('Number of Kills Because of Terrorist Incidents')
plt.show()
```


The number of deaths caused by terrorist incidents remained relatively low until the early 2000s, but increased significantly after 2001, reaching a sharp peak around 2014–2015. This spike corresponds to the rise of ISIS and the conflicts in Iraq and Syria. After 2015, the number of deaths gradually declined but stayed higher than pre-2000 levels.

### Nkill per attack for each weapon type

```python
weapon_casualties = df.groupby('weaptype1_txt')['nkill'].mean().sort_values(ascending=False)
weapon_casualties
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=weapon_casualties.values, y=weapon_casualties.index, palette="viridis")
plt.title("Average Number of Kills per Attack by Weapon Type", size=16)
plt.xlabel("Average Number of Kills", size=14)
plt.ylabel("Weapon Type", size=14)
plt.savefig('Average Number of Kills per Attack by Weapon Type')
plt.show()
```


Insight: Weapons of mass destruction such as chemical, biological, or radiological weapons result in the highest average number of deaths per attack, although they are used rarely. In contrast, explosives and firearms, while more commonly used, cause fewer deaths on average per incident. Melee and incendiary weapons tend to have the lowest fatality rates per attack.
